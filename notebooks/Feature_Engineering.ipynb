**Feature Engineering**

now,since we have evaluated and cleaned  the data and  have all the neccesary insights, we move further to keep features that are relevant for our target

```python
import pandas as pd
```

```python
data = pd.read_csv('/content/Metro_Interstate_Traffic_Volume.csv')
```

**okay now based on what insights  we have found, we will extract features**

```python
data['date_time']=pd.to_datetime(data['date_time'])
```

```python
data['hour']=data['date_time'].dt.hour
```

```python
data['weekdays']=data['date_time'].dt.dayofweek
```

```python
data['month']=data['date_time'].dt.month
```

```python
data['year']=data['date_time'].dt.year
```

```python
data['is_holiday']=data['holiday'].notna().astype(int)
```

```python
data['is_weekend']=(data['weekdays']>=5).astype(int)
```

```python
data['is_rush_hour'] = ((data['hour'] >= 6) & (data['hour'] <= 8) | (data['hour'] >= 15) & (data['hour'] <= 18  )).astype(int)
```

```python
data['heavy_rain']=(data['rain_1h'] > 2).astype(int)
```

```python
data['is_raining'] = (data['rain_1h'] > 0).astype(int)
```
